```python
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import fft, fftshift
import math
import operator
```

```python
filename = '/content/drive/MyDrive/pertemuan_3_forensika_suara/a.wav'
fs, x = wav.read(filename)
x = x[:,0]
x = (x - min(x)) / (max(x) - min(x)) * 2 - 1
x = x - np.mean(x)
t = np.arange(0,len(x)/fs,1/fs)
```

**Analisis waveform**

```python
plt.plot(t,x)
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitude')
plt.title('Waveform dari suara a.wav')
```

**Analisis Spektrum**

```python
nfft = 2 ** math.ceil(math.log(len(x),2))
nfft
```

```python
X = fftshift(fft(x,nfft))
f = np.arange(-nfft/2,nfft/2) * fs / nfft
low_lim = int(len(X)/2)
upp_lim = low_lim + 3000
plt.plot(f[low_lim:upp_lim], np.abs(X[low_lim:upp_lim]))
plt.xlabel('Frekuensi (Hz)')
plt.ylabel('Amplitude')
```

```python
index, value = max(enumerate(np.abs(X[low_lim:upp_lim])), key=operator.itemgetter(1))
```
